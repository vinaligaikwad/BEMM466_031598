# ðŸ§¹ Step 1: Data Cleaning & Preparation
This notebook merges both sheets from the Online Retail dataset, performs cleaning, and outputs a ready-to-use dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')
```

```python
# Load Excel and combine both sheets
xls = pd.ExcelFile('online_retail.xlsx')
df1 = pd.read_excel(xls, sheet_name=xls.sheet_names[0])
df2 = pd.read_excel(xls, sheet_name=xls.sheet_names[1])
df = pd.concat([df1, df2], ignore_index=True)
df.head()
```

```python
# Clean the data
df.dropna(subset=['Customer ID'], inplace=True)
df.drop_duplicates(inplace=True)
df = df[(df['Quantity'] > 0) & (df['Price'] > 0)]
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['Customer ID'] = df['Customer ID'].astype(int)
df['TotalPrice'] = df['Quantity'] * df['Price']
df.to_csv('online_retail_cleaned.csv', index=False)
df.head()
```
